<b style = 'font-size:24px;font-family:Arial;color:#E37C4D'>Naming Standards</b>
   
<p style = 'font-size:20px;font-family:Arial;color:#E37C4D'>Compute Group (CG) naming standard</p>

<p style = 'font-size:16px;font-family:Arial'>CG_<i>BusGrpName_type</i></p>

<p style = 'font-size:16px;font-family:Arial'>where <i>BusGrpName</i>  Business Group (if more than 1 CG, reflect in name)</p>
<p style = 'font-size:16px;font-family:Arial'>    <i>type</i>        Compute Group type (STD/ANL) (ie: Standard/Analytic)</p>
      
<p style = 'font-size:16px;font-family:Arial'> eg: CG_BusGrpA_STD<br>
eg: CG_BusGrpA_ANL</p>


<p style = 'font-size:20px;font-family:Arial;color:#E37C4D'>Compute Group Role (CR) naming standard</p>
<p style = 'font-size:16px;font-family:Arial'>CR_<i>BusGrpName_type</i> <br>
eg: CR_BusGrpA_STD<br>
eg: CR_BusGrpA_ANL</p>


<p style = 'font-size:20px;font-family:Arial;color:#E37C4D'>Compute Group Profile (CP) naming standard</p>
<p style = 'font-size:16px;font-family:Arial'>CP_<i>BusGrpName_type_n_size</i><br>
    where <i>BusGrpName</i>  Business Group (if more than 1 CG, reflect in name)<br>
    <i>type</i>       Compute Group type (STD/ANL)<br>
    <i>n</i>           Compute Profile idendifier (n=1-9)<br>
    <i>size</i>        Compute Cluster t-shirt size (TNY/SML/MED/LRG/1XL/2XL)<br>

eg: CP_BusGrpA_STD_1_SML<br>
eg: CP_BusGrpA_STD_2_MED<br>
eg: CP_BusGrpA_ANL_1_SML</p>

````python
# Import the minimal number of packages

import warnings
warnings.filterwarnings('ignore')

import json
import pandas as pd
import teradatasql

from teradataml import *
````

````python
# load vars json
with open('vars.json', 'r') as f:
    session_vars = json.load(f)
````


## Section 1: Environment Setup ## 
### steps overview ###

01) Create TD admin super user (sysdba) and grant privileges   [as dbc]
02) Create User hierachy & Compute Group admin (cgadmin)       [as sysdba]
03) Grant Compute Group admin (cgadmin) privileges             [as dbc]
04) Create Environment database hierachy                       [as sysdba]
05) Create user profiles                                       [as sysdba]
06) Create Roles                                               [as sysdba]
07) Grant privileges to Roles                                  [as sysdba]
08) Create Compute Groups / Compute Profiles                   [as cgadmin]

````python
# /* -------------------------------------------------------- */
# -- Perform this as DBC.
# -- Create Super User (SYSDBA). Parent of hierarchy.
# /* -------------------------------------------------------- */

dbc_name = session_vars['hierarchy']['DBC']['username']
dbc_pwd = session_vars['hierarchy']['DBC']['password']
name = session_vars['hierarchy']['SYSDBA']['username']
pwd = session_vars['hierarchy']['SYSDBA']['password']
perm = session_vars['hierarchy']['SYSDBA']['perm']
spool = session_vars['hierarchy']['SYSDBA']['spool']


with teradatasql.connect(host = session_vars['environment']['host'], 
                     user = dbc_name, 
                     password = dbc_pwd) as con:
    cur = con.cursor()
    
    qry = f'''
        CREATE USER {name} FROM DBC
          AS PERM  = {perm}          /* 800GB */
            ,SPOOL = {spool}          /* 800GB */
            ,PASSWORD = "{pwd}"
            ,ACCOUNT = ('$M')
            ,TIME ZONE = '00:00'   /* UDT */
            ,DEFAULT ROLE = ALL  ;
        '''
    try:
        cur.execute(qry)
    except Exception as e:
        # check if user exists
        if str(e.args).find("Error 5612") >= 1:
            pass
        else:
            raise
            
    
    queries = [f'''
        --Grant All privileges on SYSDBA to itself.
        GRANT all on "{name}" to "{name}" WITH GRANT OPTION;
        ''', 
              f'''
        --Grant system-level privileges to SYSDBA.
        GRANT MONRESOURCE, MONSESSION, ABORTSESSION, SETSESSRATE, SETRESRATE, REPLCONTROL, 
              CREATE PROFILE, CREATE ROLE, DROP PROFILE, DROP ROLE 
            to "{name}" WITH GRANT OPTION;
        ''', 
              f'''GRANT EXECUTE, SELECT, STATISTICS, SHOW on DBC to "{name}" WITH GRANT OPTION;''', 
              f'''GRANT UDTTYPE, UDTMETHOD on SYSUDTLIB to "{name}" WITH GRANT OPTION;''', 
              f'''GRANT SELECT, INSERT, UPDATE, DELETE on "Sys_Calendar" to "{name}" WITH GRANT OPTION;''',
              f'''GRANT SELECT, EXECUTE FUNCTION, EXECUTE PROCEDURE on TD_SYSFNLIB to "{name}" WITH GRANT OPTION;''',
              f'''GRANT SELECT, EXECUTE FUNCTION, EXECUTE PROCEDURE on TD_MLDB to "{name}" WITH GRANT OPTION;''',
              f'''GRANT SELECT, EXECUTE FUNCTION, EXECUTE PROCEDURE on syslib to "{name}" WITH GRANT OPTION;''', 
              f'''GRANT EXECUTE FUNCTION on TD_SYSFNLIB.READ_NOS to "{name}" WITH GRANT OPTION;''', 
              f'''GRANT CREATE DATASET SCHEMA on SYSUDTLIB to "{name}" WITH GRANT OPTION;''', 
              f'''GRANT SELECT on TD_METRIC_SVC to "{name}" WITH GRANT OPTION;''',
              f'''GRANT SELECT, EXECUTE FUNCTION, EXECUTE PROCEDURE on SYSSPATIAL to {name} WITH GRANT OPTION;''', 
              f'''GRANT ALL ON retail_sample_data to "{name}" WITH GRANT OPTION;''', 
              f'''
        /* Grant SYSDBA the privileges to grant others to System Roles, except TD_COMPUTE_CLUSTER_ADMIN. */
        /* This is to lockdown who may administer Computer Groups.                                        */ 
        GRANT  TD_USER_ADMIN
              ,TD_DATA_SCIENTIST
              ,TD_FLOW_COMPOSER
              ,TD_CURRENT_QUERIES_ADMIN
              ,TD_CURRENT_QUERIES_VIEW
              ,TD_QUERY_HISTORY_VIEW
              ,DS_ADMIN
              ,DS_CITIZEN
              ,DS_PROVIDER
            to {name} WITH ADMIN OPTION;
        ''', 
              f'''GRANT LOGON ON ALL TO {name} WITH NULL PASSWORD;''']
    for qry in queries:
        cur.execute(qry)
````

````python
#/* -------------------------------------------------------- */
#-- Perform this as SYSDBA.
#-- Create User hierachy
#/* -------------------------------------------------------- */

name = session_vars['hierarchy']['SYSDBA']['username']
pwd = session_vars['hierarchy']['SYSDBA']['password']

user_perm = session_vars['hierarchy']['users']['perm']
user_spool = session_vars['hierarchy']['users']['spool']

# calculate the amount of perm for each of
# the business users (.75) and support users (.25) hierarchy:

bus_perm = str(int(float(user_perm)*.75))
sup_perm = str(int(float(user_perm)*.25))


with teradatasql.connect(host = session_vars['environment']['host'], 
                     user = name, 
                     password = pwd) as con:
    cur = con.cursor()
    
    qry = f'''
        CREATE USER "Users" FROM SYSDBA
          AS PERM  = {user_perm}       /*   30GB */
            ,SPOOL = {user_spool}      /* 100GB */
            ,PASSWORD = "{pwd}"
            ,ACCOUNT = ('$M')
            ,TIME ZONE='00:00'
            ,DEFAULT ROLE=ALL;
            '''
    try:
        cur.execute(qry)
    except Exception as e:
        # check if user exists
        if str(e.args).find("Error 5612") >= 1:
            pass
        else:
            raise
    
    qry = f'''
    CREATE USER "BusinessUsers" FROM Users
      AS PERM  = {bus_perm}       /*   20GB */
        ,SPOOL = {user_spool}      /* 100GB */
        ,PASSWORD = "{pwd}"
        ,ACCOUNT = ('$M')
        ,TIME ZONE = '00:00'
        ,DEFAULT ROLE = ALL ;
    '''
    try:
        cur.execute(qry)
    except Exception as e:
        # check if user exists
        if str(e.args).find("Error 5612") >= 1:
            pass
        else:
            raise
    
    qry = f'''
    CREATE USER "SupportUsers" FROM Users
      AS PERM =  {sup_perm}      /*   10GB */
        ,SPOOL = {user_spool}      /* 100GB */
        ,PASSWORD = "{pwd}"
        ,ACCOUNT = ('$M')
        ,TIME ZONE = '00:00'
        ,DEFAULT ROLE = ALL;
        '''
    try:
        cur.execute(qry)
    except Exception as e:
        # check if user exists
        if str(e.args).find("Error 5612") >= 1:
            pass
        else:
            raise
    
    # create support users from array of support_users
    for u in session_vars['hierarchy']['users']['support_users']:
        qry = f'''
        /* Create Compute Group administrator */
        CREATE USER {u['username']} FROM SupportUsers
          AS PERM = 0       
            ,SPOOL = {user_spool}       /* 10GB */
            ,ACCOUNT = ('$M')
            ,DEFAULT ROLE = ALL
            ,PASSWORD = "{u['password']}" ;
        '''
        try:
            cur.execute(qry)
        except Exception as e:
            # check if user exists
            if str(e.args).find("Error 5612") >= 1:
                pass
            else:
                raise
    
````

````python
#/* -------------------------------------------------------- */
#-- Perform this as *** DBC ***.
#-- Grant privileges to Compute Group administrator. 
#-- NB: To lockdown who may create Compute Clusters, there is
#--     intentionally only one CG admin.
#/* -------------------------------------------------------- */

dbc_name = session_vars['hierarchy']['DBC']['username']
dbc_pwd = session_vars['hierarchy']['DBC']['password']


with teradatasql.connect(host = session_vars['environment']['host'], 
                     user = dbc_name, 
                     password = dbc_pwd) as con:
    cur = con.cursor()
    
    # iterate over all support users
    for u in session_vars['hierarchy']['users']['support_users']:
        
        queries = [f'''
        -- Compute Group super user ONLY.
        GRANT TD_USER_ADMIN, TD_COMPUTE_CLUSTER_ADMIN, TD_FLOW_COMPOSER, TD_CURRENT_QUERIES_ADMIN, DS_ADMIN TO {u['username']};
        ''', 
                  f'''GRANT SELECT ON DBC TO {u['username']};''', 
                  f'''GRANT DROP COMPUTE PROFILE to "{u['username']}" WITH GRANT OPTION;''',
                  f'''GRANT CREATE COMPUTE PROFILE to "{u['username']}" WITH GRANT OPTION;''', 
                  f'''grant CREATE COMPUTE GROUP to "{u['username']}" WITH GRANT OPTION;''',
                  f'''grant DROP COMPUTE GROUP to "{u['username']}" WITH GRANT OPTION;''', 
                  f'''GRANT LOGON ON ALL TO {u['username']} WITH NULL PASSWORD;''']
        for qry in queries:
            cur.execute(qry)
````

````python
# /* -------------------------------------------------------- */
# -- Perform this as SYSDBA.
# -- Create Environment database hierachy
# -- Root database contains a PubAuth Authorization object for
# -- Anonymous NOS access
# /* -------------------------------------------------------- */


name = session_vars['hierarchy']['SYSDBA']['username']
pwd = session_vars['hierarchy']['SYSDBA']['password']

with teradatasql.connect(host = session_vars['environment']['host'], 
                     user = name, 
                     password = pwd) as con:
    cur = con.cursor()
    
    # Create root database
    qry = f'''
    CREATE DATABASE {session_vars['hierarchy']['repositories']['name']} from {name} 
      AS PERM = {session_vars['hierarchy']['repositories']['perm']} ;
      '''
    try:
        cur.execute(qry)
    except Exception as e:
        # check if user exists
        if str(e.args).find("Error 5612") >= 1:
            pass
        else:
            raise
    
    # Create root Public Auth object with NULL username and password
    qry = f'''REPLACE AUTHORIZATION {session_vars['hierarchy']['repositories']['name']}.PubAuth
            user '' password '';'''
    cur.execute(qry)
    
    # Create child databases
    for d in session_vars['hierarchy']['repositories']['databases']:
        qry = f'''
        CREATE DATABASE {d['name']} from Repositories  AS
          DEFAULT STORAGE = {d['default_storage']} OVERRIDE ON ERROR,
          PERM = {d['perm']};
        '''
        try:
            cur.execute(qry)
        except Exception as e:
            # check if user exists
            if str(e.args).find("Error 5612") >= 1:
                pass
            # check if object exists - "perm space"
            elif str(e.args).find("Error 3541") >= 1:
                pass
            else:
                raise
        
        qry = f'''GRANT Select on DBC to {d['name']} with grant option;'''
        cur.execute(qry)
        
        qry = f'''GRANT Select on Sys_Calendar to {d['name']} with grant option;'''
        cur.execute(qry)
        
        qry = f'''GRANT Select on TD_METRIC_SVC to {d['name']} with grant option;'''
        cur.execute(qry)
        
        qry = f'''GRANT EXECUTE on retail_sample_data.demo_Auth_NOS to "{d['name']}" WITH GRANT OPTION;'''
        cur.execute(qry)
        
        # Grant execute to the PubAuth object
        qry = f'''GRANT EXECUTE on {session_vars['hierarchy']['repositories']['name']}.PubAuth to "{d['name']}" WITH GRANT OPTION;'''
        cur.execute(qry)
        
````

````python
# /* -------------------------------------------------------- */
# -- Perform this as SYSDBA.
# -- Create Profiles
# /* -------------------------------------------------------- */

name = session_vars['hierarchy']['SYSDBA']['username']
pwd = session_vars['hierarchy']['SYSDBA']['password']

user_spool = session_vars['hierarchy']['users']['spool']

with teradatasql.connect(host = session_vars['environment']['host'], 
                     user = name, 
                     password = pwd) as con:
    cur = con.cursor()
    
    qry = f'''
        CREATE PROFILE "P_BusGrpA" AS
        SPOOL={user_spool}      /* 100GB */
        TEMPORARY=1E8   /* 100MB */
        ACCOUNT=NULL
        DEFAULT DATABASE=NULL ;
        '''
    try:
        cur.execute(qry)
    except Exception as e:
        # check if user exists
        if str(e.args).find("Error 5652") >= 1:
            pass
        else:
            raise
    
    qry = f'''
        CREATE PROFILE "P_SupportUsers" AS
        SPOOL={user_spool}      /* 100GB */
        TEMPORARY=1E8   /* 100MB */
        ACCOUNT=NULL
        DEFAULT DATABASE=NULL ;
        '''
    try:
        cur.execute(qry)
    except Exception as e:
        # check if user exists
        if str(e.args).find("Error 5652") >= 1:
            pass
        else:
            raise
````

````python
# /* -------------------------------------------------------- */
# -- Perform this as SYSDBA.
# -- Create Roles for Compute Groups admin, General TD access, NOS, etc.
# /* -------------------------------------------------------- */

name = session_vars['hierarchy']['SYSDBA']['username']
pwd = session_vars['hierarchy']['SYSDBA']['password']

with teradatasql.connect(host = session_vars['environment']['host'], 
                     user = name, 
                     password = pwd) as con:
    cur = con.cursor()
    
    for d in session_vars['hierarchy']['compute_group_roles']:
        
        qry = f'''create role {d['name']};'''
        try:
            cur.execute(qry)
        except Exception as e:
            # check if user exists
            if str(e.args).find("Error 5612") >= 1:
                pass
            else:
                raise
    
    queries = ['''
    -- Create a Teradata general admin role 
    create role R_TD_Admin; ''', 
               '''
    -- Create Roles for general TD access rights 
    create role R_TD_General;''', 
              '''
    -- Create Roles for NOS access rights 
    create role R_NOS;''', 
              '''
    -- Create ANALYTICS Role
    create role R_Analytics_1;''', 
              '''
    -- Create Demo Role (Read only)
    create role R_Demo_R;''', 
              '''
    -- Create Demo Role (Read/Write)
    create role R_Demo_RW;''']
    
    for qry in queries:
        try:
            cur.execute(qry)
        except Exception as e:
            # check if user exists
            if str(e.args).find("Error 5612") >= 1:
                pass
            else:
                raise
````

````python
# /* -------------------------------------------------------- */
# -- Perform this as cgadmin
# -- Create Compute Groups / Compute Profiles
# /* -------------------------------------------------------- */

# /* Learnings: 
# --  1) Some of these statements do not run under ODBC or .NET (only JDBC) 
# --  2) If you create a Compute Profile "INITIALLY_SUSPENDED", it goes through its
# --     states of "Deploying", "Hibernating", "Hibernated". This takes about 10 mins.
# --     Once it's in the "Hibernated" (suspend) state, currently we are not able to
# --     "Resume" it. Therefore, start/stop the CG by Create/Drop the Compute Profile.
# */

name = session_vars['hierarchy']['users']['support_users'][0]['username']
pwd = session_vars['hierarchy']['users']['support_users'][0]['password']

with teradatasql.connect(host = session_vars['environment']['host'], 
                     user = name, 
                     password = pwd) as con:
    cur = con.cursor()
    
    # iterate through the groups and profiles
    for cg in session_vars['hierarchy']['compute_groups']:
        
        qry = f'''CREATE COMPUTE GROUP {cg['name']} USING QUERY_STRATEGY('{cg['type']}');'''
        try:
            cur.execute(qry)
        except Exception as e:
            # check if group exists
            if str(e.args).find("Error 4822") >= 1:
                pass
            else:
                raise
        
        # iterate over each child profile in the group
        for cp in cg['profiles']:
            
            qry = f'''
            /* This profile is configured to be available from 9am to 9pm EDT. */
            CREATE COMPUTE PROFILE {cp['name']} IN COMPUTE GROUP {cg['name']}
            ,INSTANCE = {cp['size']}
            ,INSTANCE TYPE = {cp['type']}   
             USING
                 MIN_COMPUTE_COUNT ({cp['min']})
                 MAX_COMPUTE_COUNT ({cp['max']})
                 SCALING_POLICY ('STANDARD')
                 START_TIME ('0 13 * * MON-FRI')
                 END_TIME ('0 01 * * MON-SAT')
                 COOLDOWN_PERIOD (30) ;
            '''
            try:
                cur.execute(qry)
            except Exception as e:
                # check if profile exists
                if str(e.args).find("Error 4826") >= 1:
                    pass
                else:
                    raise
````

````python
# /* -------------------------------------------------------- */
# -- Perform this as SYSDBA.
# -- Create end-users
# -- Recommendation: Create new users under "BusinessUsers"
# /* -------------------------------------------------------- */

name = session_vars['hierarchy']['SYSDBA']['username']
pwd = session_vars['hierarchy']['SYSDBA']['password']
user_perm = session_vars['hierarchy']['users']['perm']

# split up perm space based on the number of users:
# business users are .75 of total user perm
# divide this by the number of users in business users

ind_user_perm = str(int((float(user_perm)*.75) / len(session_vars['hierarchy']['users']['business_users'])))



with teradatasql.connect(host = session_vars['environment']['host'], 
                     user = name, 
                     password = pwd) as con:
    cur = con.cursor()
    
    for u in session_vars['hierarchy']['users']['business_users']:
        
        qry = f'''
        CREATE USER {u['username']} FROM BusinessUsers
         AS PERM =  {ind_user_perm} 
           ,Spool = 0    /* Spool is defined in the Profile */
           ,Profile = "P_BusGrpA"
           ,PASSWORD = "{u['password']}"
           ,ACCOUNT = ('$M')
           ,TIME ZONE = '00:00'
           ,DEFAULT ROLE = ALL
           ,COMPUTE GROUP = {u['compute_group']};
        '''
        try:
            cur.execute(qry)
        except Exception as e:
            # check if user exists
            if str(e.args).find("Error 5612") >= 1:
                pass
            else:
                raise
        
        qry = f'''GRANT LOGON ON ALL TO {u['username']} WITH NULL PASSWORD;'''
        cur.execute(qry)
        
        qry = f'''GRANT TD_DATA_SCIENTIST TO {u['username']};'''
        cur.execute(qry)
        
        qry = f'''GRANT TD_FLOW_COMPOSER TO {u['username']};'''
        cur.execute(qry)
        
        qry = f'''GRANT DS_ADMIN TO {u['username']};'''
        cur.execute(qry)
        
        qry = f'''GRANT EXECUTE PROCEDURE ON {u['username']} TO {u['username']};'''
        cur.execute(qry)

        qry = f'''GRANT CREATE PROCEDURE ON {u['username']} TO {u['username']};'''
        cur.execute(qry)


````
