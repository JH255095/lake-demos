<p style = 'font-size:24px;font-family:Arial;color:#00233C'><b>Create JWT tokens to call VantageCloud Lake Query Service APIs</b></p>

<br>

<p style = 'font-size:16px;font-family:Arial;color:#00233C'>VantageCloud Lake exposes a set or REST APIs to manage environments. To call VantageCloud Lake REST APIs such as Query Service you need a valid JWT token. This quickstart demonstrates how to create a JWT token and how to use it to call VantageCloud Lake REST APIs.</p>

<p style = 'font-size:16px;font-family:Arial;color:#00233C'>This notebook is based off of a getting started guide posted <a href = 'https://developers.teradata.com/quickstarts/vantagecloud-lake/create-jwt-token-for-api-calls/'>here</a> on <a href = 'https://developers.teradata.com'>https://developers.teradata.com</a>.</p>

<p style = 'font-size:16px;font-family:Arial;color:#00233C'><b>Steps</b></p>
<ol style = 'font-size:16px;font-family:Arial;color:#00233C'>

    <li>Create a JWT for API authentication</li>
    <li>Call the API using the JWT</li>
    </ol>

<hr>
<p style = 'font-size:24px;font-family:Arial;color:#00233C'><b>Step 1 - Create PAT, Keys, and obtain org ID</b></p>

<p style = 'font-size:16px;font-family:Arial;color:#00233C'>For purposes of enhanced security, the VantageCloud Lake API endpoints require the use of a JWT (Java Web Token) that is constructed using a "Personal Access Token" and a private/public key pair.  The Personal Access Token is a one-time-visible key that has a set expiration date and can be revoked manually at any time by the user or administrators.  The key pair is an RSA-256 encrypted key that can be generated using VantageCloud Lake console, or users can upload their own public key file.</p>

<p style = 'font-size:16px;font-family:Arial;color:#00233C'><b>The process is as follows</b></p>
<ol style = 'font-size:16px;font-family:Arial;color:#00233C'>
    <li>Log into Console with the <b>database user</b> who is going to make the API calls</li>
    <li>Generate the PAT and copy the value</li>
    <li>Generate the keypair, download the private key</li>
    <li>Access the copy the site's organization ID</li>
    </ol>

<p style = 'font-size:24px;font-family:Arial;color:#00233C'><b>Step 2 - Create the JWT</b></p>

<p style = 'font-size:16px;font-family:Arial;color:#00233C'>A valid JWT can be constructed using any one of a number of languages and tools.  Examples here include python and bash (shell scripting) - please see the getting started guide posted <a href = 'https://developers.teradata.com/quickstarts/vantagecloud-lake/create-jwt-token-for-api-calls/'>here</a> on <a href = 'https://developers.teradata.com'>https://developers.teradata.com</a> for more example code in additional languages.</p>

<p style = 'font-size:16px;font-family:Arial;color:#00233C'><b>The process is as follows</b></p>
<ol style = 'font-size:16px;font-family:Arial;color:#00233C'>
    <li>Construct the token header</li>
    <li>Construct the token payload</li>
    <li>Encode the header and payload data</li>
    <li>Digitally sign the encoded header and payload with the user's private key</li>
    <li>Encode the signature</li>
    <li>Assemble these values into the JWT string</li>
    </ol>
    
<hr>

<p style = 'font-size:18px;font-family:Arial;color:#00233C'><b>Python package imports</b></p>

<p style = 'font-size:16px;font-family:Arial;color:#00233C'>Import the required packages.</p>

````python
import base64
import json
import time
import requests
import datetime
import pandas as pd

from cryptography.hazmat.primitives import hashes
from cryptography.hazmat.primitives.asymmetric import padding
from cryptography.hazmat.primitives.serialization import load_pem_private_key
````

<hr>

<p style = 'font-size:18px;font-family:Arial;color:#00233C'><b>Provide the required inputs</b></p>

<p style = 'font-size:16px;font-family:Arial;color:#00233C'>Modify the values below to match the specific user's environment.  For this demonstration, some values have been provided using a "vars.json" input file (database user, PAT, and path to pem file).  Overwrite these if necessary, and provide the required org_id, database environment name, and API Hostname:</p>

````python

# load vars json
with open('../../vars.json', 'r') as f:
    session_vars = json.load(f)

# Database login information
username = session_vars['hierarchy']['users']['business_users'][0]['username']

# API Authentication information
pat_token = session_vars['hierarchy']['users']['business_users'][0]['pat_token']
pem_file = session_vars['hierarchy']['users']['business_users'][0]['key_file']

org_id = '3...7'
db_name = 'my_environment'
hostname = 'mydomain.innovationlabs.teradata.com'
````

<hr>

<p style = 'font-size:24px;font-family:Arial;color:#00233C'><b>Step 2 - Construct the JWT</b></p>

<p style = 'font-size:16px;font-family:Arial;color:#00233C'>Two functions are defined in the code below.  The first ("base64_url_encode") will return URL- and filesystem-safe Base64 encoding of an input string.  The second function ("create_jwt") will return the JWT as a Base64-encoded string.  For detailed information on JWT structure and fields, please see <a href = 'https://en.wikipedia.org/wiki/JSON_Web_Token'>online resources</a>.  Briefly, the JWT contains:</p>

<ol style = 'font-size:16px;font-family:Arial;color:#00233C'>
    <li><b>A header</b>, which can contain various metadata information about the token.  For VantageCloud Lake, the header needs to include the signing algorithm ("alg"), the key id ("kid"), and token type ("typ").  The <b>kid</b> is the key name that was used when the keypair was generated in VantageCloud Lake Console.  Most of the time, this will be the key filename (less extension) as well - this is the convention used in this demo</li>
    <li><b>Payload</b> which contains verifiable information about the user, service permissions, expiration, etc.  For VantageCloud Lake, additional payload elements include the Org ID collected above, the Personal Access Token, username, and expiration timestamp. <b>Note</b> one can extend the expiration of the JWT by modifying the value in the "exp" object</li>
    </ol>

<p style = 'font-size:16px;font-family:Arial;color:#00233C'>The JWT is constructed by concatenating the Base64-encoded header, payload, and a <b>signature</b> which is a cryptographical signature of the header and payload.  This combination allows the authenticating service to use the user's public key to validate the consistency of the header and payload values.</p>

````python
def base64_url_encode(data):
    return base64.urlsafe_b64encode(data).rstrip(b'=').decode('utf-8')

def create_jwt(org_id, pat, user, key):
    # Define the header
    header = {
        "alg": "RS256",
        "kid": key.split('.')[0],
        "typ": "JWT"
    }
    header_json = json.dumps(header)
    header_encoded = base64_url_encode(header_json.encode('utf-8'))

    # Define the payload
    payload = {
        "aud": ["td:service:authentication"],
        "exp": int(datetime.datetime.now().timestamp()) + 3600,
        "iat": int(datetime.datetime.now().timestamp()),
        "iss": "wiki",
        "org_id": org_id,
        "multi-use": True,
        "pat": pat,
        "sub": user
    }
    payload_json = json.dumps(payload)
    payload_encoded = base64_url_encode(payload_json.encode('utf-8'))

    # Load the private key
    with open(key, 'rb') as key_file:
        private_key = load_pem_private_key(key_file.read(), password=None)

    # Create the signature
    data = f"{header_encoded}.{payload_encoded}".encode('utf-8')
    signature = private_key.sign(
        data,
        padding.PKCS1v15(),
        hashes.SHA256()
    )
    signature_encoded = base64_url_encode(signature)

    # Construct the JWT
    jwt = f"{header_encoded}.{payload_encoded}.{signature_encoded}"

    return jwt
````

<hr>

<p style = 'font-size:18px;font-family:Arial;color:#00233C'><b>Call the functions</b></p>

<p style = 'font-size:16px;font-family:Arial;color:#00233C'>The create_jwt function takes the org_id, PAT value, username, and private key filename as inputs, and returns a string containing the valid JWT token.</p>

````python
jwt = create_jwt(org_id = org_id, pat = pat_token, user = username, key = pem_file)
print(jwt)
````

<hr>

<p style = 'font-size:24px;font-family:Arial;color:#00233C'><b>Step 3 - Issue the API request</b></p>

<p style = 'font-size:16px;font-family:Arial;color:#00233C'>Construct a valid REST query to invoke the Query Service.  This HTTP request can be any of the supported API calls documented <a href = 'https://downloads.teradata.com/api/teradata_query_service'>here</a>.  For this demonstration, a simple POST is constructed to the /queries endpoint to issue a SQL statement.  This request consists of:</p>

<ul style = 'font-size:16px;font-family:Arial;color:#00233C'>
    <li><b>URL</b>.  This is the API endpoint.  Hostname is dependent on the user's environment and is captured here in the "hostname" variable.  The path requires the database "environment name", but the remainder is fixed for the method invoked (/queries) and represents the Query Service endpoint</li>
    <li><b>Headers</b>.  The HTTP header requires three fields; "Accept", "Content-Type", and "Authorization", where the JWT is passed as part of the "Authorization" value, following the literal "Bearer{space}{JWT}"</li>
    <li><b>Postdata</b>.  A JSON-formatted string that contains the query.  The documentation linked above provides detailed information on the additional objects that can be added to the postdata string</li>
    </ul>

````python
# Make the HTTP request


url = f'https://{hostname}/api/tdrest/systems/{db_name}/queries'
headers = {
    'Accept': 'application/json',
    'Authorization': f'Bearer {jwt}',
    'Content-Type': 'application/json'
}

req_body = json.dumps({"query": "SEL * FROM dbc.dbcinfo", "logMech": "JWT"})
````

<hr>

<p style = 'font-size:18px;font-family:Arial;color:#00233C'><b>Invoke the request</b></p>

<p style = 'font-size:16px;font-family:Arial;color:#00233C'>Using the Python requests package, issue a POST, pass the URL, headers, and postdata.  Print the response status code (200 for success) and text.  In this example, the response text is a json-formatted string that contains the query resultset and metadata.</p>

````python
response = requests.post(url, headers=headers, data=req_body)
print("Response Code:", response.status_code)
print("Response Body:", response.text)
````

<hr>

<p style = 'font-size:18px;font-family:Arial;color:#00233C'><b>Format results</b></p>

<p style = 'font-size:16px;font-family:Arial;color:#00233C'>The Python ecosystem provides extensive tooling to parse and manipulate json-formatted strings.  This example uses the json and pandas packages to convert the resultset into a pandas DataFrame.</p>

````python
# use json loads to convert the response text to a json object
res = json.loads(response.text)

# pass the resultset object to the DataFrame constructor
df = pd.DataFrame(res['results'][0]['data'])

df
````

<p style = 'font-size:24px;font-family:Arial;color:#00233C'><b>Bonus - use bash</b></p>

<p style = 'font-size:16px;font-family:Arial;color:#00233C'>Using the same values as above, construct the JWT using bash syntax.  Issue the request using curl.  Be sure to maintain the proper quoting and escape characters as illustrated below.</p>
<ul style = 'font-size:16px;font-family:Arial;color:#00233C'>
    <li><b>kid header value</b>.  Replace this with the name (less file extension if using this convention) of the private key file</li>
    <li><b>payload values</b>. Replace the org_id, pat, and sub values - ensure the proper escaped double-quotes are maintained</li>
    <li><b>signature</b>.  Ensure the openssl command -sign argument points to the fully-qualified or relative path to the private key file</li>
 </ul>

```bash
#!/usr/bin/sh
# define the header, the kid needs to be the same as the key name defined in the console, i.e. `primary`
header=$(echo -n '{"alg":"RS256", "kid":"qs", "typ":"JWT"}')


# define the payload, here is the description of different params from the wiki
# exp       - Expiration claim. This is a UTC epoch and recommended to be at 
#             least 30 seconds into the future. You can use Epoch Converter 
#             website: https://www.epochconverter.com/
# iat       - Issued At claim. This is a UTC epoch. Set to the current time
# iss       - The issuer. Can be anything and will be recorded in the logs
# multi-use - Determines if the token can be used more than once.
# org_id    - org_id (you can get it from the console in dev mode)
# pat       - The pat to be exchanged (retrieved from the console)
# sub       - The owner of the PAT. This must match the owner specified when 
#             creating the pat. (the name of the user that you used to log in 
#             to the console)

payload=$(echo -n "{
  \"aud\": [\"td:service:authentication\"],
  \"exp\": $(($(date +%s) + 3600)),
  \"iat\": $(date +%s),
  \"iss\": \"wiki\",
  \"org_id\": \"38...7\",
  \"multi-use\": true,
  \"pat\": \"g...6\",
  \"sub\": \"data_engineer\"
}")

export header=$(echo -n "$header" | base64 | tr -d '\n=' | tr -- '+/' '-_' )
export payload=$(echo -n "$payload" | base64 | tr -d '\n=' | tr -- '+/' '-_' )
export signature=$(echo -n "$header.$payload" | openssl dgst -sha256 -binary -sign qs.pem | openssl enc -base64 | tr -d '\n=' | tr -- '+/' '-_' )
export jwt=$(echo -n "$header.$payload.$signature")

sleep 2

curl -X POST 'https://mydomain.innovationlabs.teradata.com/api/tdrest/systems/shared_demo/queries' \
  --header 'Accept: application/json' \
  --header "Authorization: Bearer $jwt" \
  --data-raw '{"query":"SEL * FROM dbc.dbcinfo","logMech":"JWT"}'
 ```
